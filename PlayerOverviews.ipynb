```python
import pandas as pd
```

```python
years = [str(year) for year in range(1953,1974)]
impacts = pd.DataFrame()
for year in years:
    impact = pd.read_excel('InformedImpact/'+year+'impactInformed.xlsx')
    impact['year']=str(year)
    if len(impacts) == 0:
        impacts = impact.copy()
    else:
        impacts=pd.concat([impacts, impact], axis=0)
impacts
```
